```python
import pandas

config = pandas.read_json("../config.json", typ="series")
config.LUCAS_RDATA
```

```python
import rpy2.robjects as robjects

robjects.r['load'](config.LUCAS_RDATA)
```

```python
Rdf = robjects.r('LUCAS.SOIL')
Rdf
```

```python
colnames = list(Rdf.colnames)
cols_to_extract = ["sample.ID", "ID", "clay", "silt", "sand", "pH.in.H2O", "OC", "N", "P", "K", "CEC", "CaCO3"]
cols_dict = {key: colnames.index(key) for key in cols_to_extract}
cols_dict
```

```python
df = pandas.DataFrame({key: list(Rdf[cols_dict[key]]) for key in cols_to_extract}).set_index("ID")
df
```

```python
spc = Rdf[3]
spc
```

```python
spc_df = pandas.DataFrame({key: list(spc[i]) for i, key in enumerate(list(spc.colnames))})
spc_df["ID"] = df.index
spc_df.set_index("ID", inplace=True)
spc_df
```
